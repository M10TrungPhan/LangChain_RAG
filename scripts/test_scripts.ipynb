```python
def get_embedding_text(text:str):
    # text = "Làm thế nào để tạo điều kiện đảm bảo thoát nước tốt và hạn chế ngập úng cho cây sầu riêng?"
    output = rdrsegmenter.word_segment(text)
    input_ids = torch.tensor([tokenizer.encode(output,padding='max_length', truncation=True, max_length=256)])
    with torch.no_grad():
        features = phobert(input_ids)
    last_hidden_state, pooler_output = features[0], features[1]
    return torch.mean(last_hidden_state, dim=1)
```
