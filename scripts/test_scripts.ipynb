```python
import requests
import re
import markdownify 
from bs4 import BeautifulSoup
questions = []
getable = True
page = 1
limit = 100
```

```python
def split_by_punctuation(input_string):
    tokens = [input_string] + re.split(r"[.?!]", input_string)
    tokens = [re.sub(r"[.?!]$", "", token) for token in tokens]
    tokens = [token.strip() for token in tokens if token.strip()]

    return tokens
```

```python
while getable:
    response = requests.get(
        f"https://panel.hainong.vn/api/v2/knowledge_handbooks?page={page}&limit={limit}"
    )
    if response.ok:
        data = response.json().get("data")
        # print(data)
        if len(data) > 0:
            number_doc =0
            for knowledge_handbook in data:
                
                # questions += split_by_punctuation(knowledge_handbook.get("question"))
                html_content = knowledge_handbook.get("answer")
                
                question_content =  markdownify.markdownify(knowledge_handbook.get("question"), heading_style="ATX") 

                file_name = 'knowledge_handbooks_' + question_content.replace(" ","_"). \
                                                                    replace("/","_"). \
                                                                    replace(",","_"). \
                                                                    replace("?","") + '.md'
                answer_content =  markdownify.markdownify(html_content, heading_style="ATX")
                content_output = question_content + "\n" +answer_content
                print(content_output)
                with open("E:/TrungPhanADVN/Code/LangChain_RAG/app/api/data/training_data/" + file_name, 'w') as f:
                    f.write(content_output)
                number_doc+=1
                if number_doc == 50:
                    break
                print("______________________________________________________________________________________________________________________________")                
        # else:
        #     getable = False
    # else:
    #     getable = False
    break
# print(questions)
    # page += 1
```
